### 데이터 확인하기

```python
import os

# 주어진 경로 내의 .tfrec 파일들을 리스트로 가져오는 함수
def get_tfrec_files(path):
    return [os.path.join(path, filename) for filename in os.listdir(path)]

# 각 경로에서 .tfrec 파일들을 배열에 담기
file_path1 = '/kaggle/input/tpu-getting-started/tfrecords-jpeg-192x192/train'
file_path2 = '/kaggle/input/tpu-getting-started/tfrecords-jpeg-224x224/train'

tfrec_files1 = get_tfrec_files(file_path1)
tfrec_files2 = get_tfrec_files(file_path2)

# .tfrec 파일 리스트 출력
print("tfrecords-jpeg-192x192:")
print(tfrec_files1[3:])

print("\ntfrecords-jpeg-224x224:")
print(tfrec_files2[3:])
```

```python
# 이미지 출력 함수
def show_image(image):
    decoded_image = tf.image.decode_jpeg(image)
    plt.imshow(decoded_image)
    plt.axis('off')
    plt.show()

# TFRecord 파일에서 데이터를 읽고 파싱하는 함수
def read_tfrecord(serialized_example):
    feature_description = {
        'image': tf.io.FixedLenFeature([], tf.string),
        'class': tf.io.FixedLenFeature([], tf.int64),
        'id': tf.io.FixedLenFeature([], tf.string),
    }
    example = tf.io.parse_single_example(serialized_example, feature_description)
    return example['id'], example['class'], example['image']

def count_tfrecords(filepaths):
    count = 0
    for filepath in filepaths:
        dataset = tf.data.TFRecordDataset(filepath)
        for _ in dataset:
            count += 1
    return count

# 파일 경로에서 tfrec 파일을 읽는 함수
def read_top_n_tfrecords(filepaths, n=1):
    for filepath in filepaths[:n]:
        dataset = tf.data.TFRecordDataset(filepath)
        total_records = count_tfrecords([filepath]) # 파일의 레코드 수를 가져옵니다.
        print(f"Dataset Length: {total_records}")
        print(f"Reading {filepath}")
        for raw_record in dataset.take(3):  # 각 파일에서 상위 3개의 레코드만 읽습니다.
            id, label, image = read_tfrecord(raw_record)
            print(f"ID: {id.numpy().decode('utf-8')}, Label: {label.numpy()}, Image shape: {tf.image.decode_jpeg(image).shape}")
            show_image(image) # 이미지 출력

# 192x192 크기의 이미지를 가진 상위 2개의 tfrec 파일 읽기
print("=== tfrecords-jpeg-192x192 ===")
read_top_n_tfrecords(tfrec_files1)

print('*' * 80)

# 224x224 크기의 이미지를 가진 상위 2개의 tfrec 파일 읽기
print("\n=== tfrecords-jpeg-224x224 ===")
read_top_n_tfrecords(tfrec_files2)
```

### 데이터 전처리

```python
# 주어진 경로의 모든 TFRecord 파일을 읽어 데이터셋을 생성하는 함수
def load_dataset(folder_path):
    tfrec_files = [os.path.join(folder_path, fname) for fname in os.listdir(folder_path) if fname.endswith('.tfrec')]
    dataset = tf.data.TFRecordDataset(tfrec_files)
    dataset = dataset.map(read_tfrecord)
    return dataset

folder_paths = [
    '/kaggle/input/tpu-getting-started/tfrecords-jpeg-192x192/',
    '/kaggle/input/tpu-getting-started/tfrecords-jpeg-224x224/',
    '/kaggle/input/tpu-getting-started/tfrecords-jpeg-331x331/',
    '/kaggle/input/tpu-getting-started/tfrecords-jpeg-512x512/'
]

test_datasets = []
train_datasets = []
val_datasets = []

for folder_path in folder_paths:
    test_folder = os.path.join(folder_path, 'test')
    train_folder = os.path.join(folder_path, 'train')
    val_folder = os.path.join(folder_path, 'val')
    
    test_datasets.append(load_dataset(test_folder))
    train_datasets.append(load_dataset(train_folder))
    val_datasets.append(load_dataset(val_folder))
    
# 각각의 데이터셋을 하나로 묶기
merged_test_dataset = test_datasets[0].concatenate(test_datasets[1]).concatenate(test_datasets[2]).concatenate(test_datasets[3])
merged_train_dataset = train_datasets[0].concatenate(train_datasets[2]).concatenate(train_datasets[1]).concatenate(train_datasets[3])
merged_val_dataset = val_datasets[0].concatenate(val_datasets[1]).concatenate(val_datasets[2]).concatenate(val_datasets[3])
```

```python
unique_labels = set()  # 중복을 허용하지 않는 set 데이터 구조를 사용

# 라벨만 추출
for _, label_tensor, _ in merged_train_dataset:
    label = label_tensor.numpy()
    unique_labels.add(label)  # set에 추가

print(f"Unique labels in the dataset: {unique_labels}")
print(f"Total number of unique labels: {len(unique_labels)}")
```

```python
for item in merged_train_dataset.take(3):
    id_tensor, label_tensor, image_tensor = item
    print(f"ID: {id_tensor.numpy().decode('utf-8')}")
    print(f"Label: {label_tensor.numpy()}")
    print(f"Image shape: {tf.image.decode_jpeg(image_tensor).shape}")
    show_image(image_tensor)
```

```python
def resize_image(id_tensor, label_tensor, image_tensor):
    # JPEG 형식의 이미지 데이터를 디코딩
    image = tf.image.decode_jpeg(image_tensor)
    # 이미지를 192x192로 리사이즈
    resized_image = tf.image.resize(image, [192, 192])
    return id_tensor, label_tensor, resized_image

# 각 데이터셋에 대해 리사이즈를 적용
resized_train_dataset = merged_train_dataset.map(resize_image)
resized_val_dataset = merged_val_dataset.map(resize_image)
resized_test_dataset = merged_test_dataset.map(resize_image)
```

### 모델링

```python
# 간단한 CNN 모델 정의
def create_model():
    model = tf.keras.Sequential([
        tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(192, 192, 3)),
        tf.keras.layers.MaxPooling2D((2, 2)),
        tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
        tf.keras.layers.MaxPooling2D((2, 2)),
        tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
        tf.keras.layers.MaxPooling2D((2, 2)),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.Dense(104, activation='softmax') 
    ])
    return model

# 모델 컴파일
model = create_model()
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.summary()
```

```python
# 데이터셋에서 레이블과 이미지 분리
def separate_labels(id_tensor, label_tensor, image_tensor):
    return image_tensor, label_tensor

train_dataset = resized_train_dataset.map(separate_labels)
val_dataset = resized_val_dataset.map(separate_labels)
```

```python
# 모델 훈련
history = model.fit(train_dataset.batch(32),
                    validation_data=val_dataset.batch(32),
                    epochs=10)
```

```python
def only_image(*args):  # 가변 인자를 사용하여 여러 형태의 데이터셋에 대응
    image_tensor = args[1]  # 두 번째 인자가 이미지 데이터라고 가정
    image = tf.image.decode_jpeg(image_tensor)
    resized_image = tf.image.resize(image, [192, 192])
    return resized_image  # 이미지만 반환

# 테스트 데이터셋에 적용
resized_test_dataset = merged_test_dataset.map(only_image)

# 모델 평가
test_loss, test_accuracy = model.evaluate(resized_test_dataset.batch(32))
print(f"Test accuracy: {test_accuracy * 100:.2f}%")
```

### csv 파일로 만들기

```python
# 예측을 위한 데이터셋 배치 설정
test_dataset_batched = test_dataset.batch(32)

# 예측 수행
predictions = model.predict(test_dataset_batched)

# 예측 라벨 추출 (가장 확률이 높은 클래스의 인덱스를 선택)
predicted_labels = np.argmax(predictions, axis=1)

# 이미지 ID를 담을 리스트
image_ids = []

# 테스트 데이터셋에서 ID 추출
for image_id, _, _ in test_dataset:
    image_ids.append(image_id.numpy().decode("utf-8"))

# DataFrame 생성
df = pd.DataFrame({
    'id': image_ids,
    'label': predicted_labels
})

# CSV 파일로 저장
df.to_csv('sample_submission.csv', index=False)
```
